```python
Options_df = pd.read_html('https://finance.yahoo.com/quote/SPY/options?p=SPY')[0]
Options_df.head()
```

```python
Options_df.dtypes
```

```python
Options_df['Last Trade Date'] = Options_df['Last Trade Date'].map(lambda x:x.split(' ')[0])
```

```python
Options_df = Options_df.rename(columns={"Last Trade Date": "Date"})
Options_df = Options_df.sort_values(by='Date', ascending=False)

Options_df['Date'] = pd.to_datetime(Options_df['Date'])
begin_date = '2020-01-27'
Options_df = Options_df.loc[(Options_df['Date'] >= begin_date)]
Options_df
```

```python
historical_price_df = pd.read_csv('SPY.csv')
historical_price_df = historical_price_df.sort_values(by='Date', ascending=False)
historical_price_df['Date'] = pd.to_datetime(historical_price_df['Date'])
most_one_week_df = historical_price_df.loc[(historical_price_df['Date'] >= begin_date)]
most_one_week_df
```

```python
Stock_Option_df = pd.merge(Options_df, most_one_week_df, on='Date', how='left')
```
