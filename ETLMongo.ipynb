```python

#--------------------------------------------------------------------------------------------------
# .                                INSERT MANY IN MONGODB
#  WE CAN ADD ADDITIONAL VARIABLES TO THIS DICT
#  FOR BREVITY AND SIMPLICITY, I ADDED THE MERGED DATA OF STOCK  PRICE + STRIKE PRICE + THE CONTRACT
#                                                                 PIRUZ ALEMI FEB 1ST, 2020
#  FOR A "SIMILAR" INSERTION IN AN SQL DATA BASE, INSTEAD OF MONGODB SEE PIRUZ_JHONPING PROJECT
#  UNDER SAME DEPOSITORY" ETL_CHALLENGE
# --------------------------------------------------------------------------------------------------
option_dict_list = []

for i in range(5):
    option_dict = {'Contract Name': spyDict_data['Contract Name'][i],
     'Strike Price': spyDict_data['Strike Price'][i],
     'Stock Price': spyDict_data['Stock Price'][i]
    }
    option_dict_list.append(option_dict)
# Insert a document into the 'stock_option' collection
db.stock_option.insert_many(option_dict_list)


# make a list of dictionaries 
# use insert many
```
